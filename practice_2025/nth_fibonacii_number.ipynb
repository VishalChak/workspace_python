```python
## Fibonacii number

# formula :   f(n) = f(n-1) + f(n-2)
# condtion : f(1)=f(2) =1

# time complexity O(2^n)
```

```python
def fibonacii(n):
    if n < 1:
        return 0
    if n == 1 or n ==2:
        return 1
    else:
        return fibonacii(n-1) + fibonacii(n-2) 
```

```python
print(fibonacii(-1))
```
