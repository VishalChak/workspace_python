```python
import pandas as pd
```

```python
df = pd.read_csv("../train_.csv")
```

```python
print(df.shape)
df.describe()
```

```python
df_c = pd.read_csv("../train_churn_labels.csv")
```

```python
df_c.Label.value_counts()
```

```python
df['label'] = df_c['Label']
```

```python
class_A_sample = df[df['label'] == -1].sample(n=50, random_state=42)
class_B_sample = df[df['label'] == 1].sample(n=5, random_state=42)

# Combine the samples
balanced_df = pd.concat([class_A_sample, class_B_sample], ignore_index=True)

# Optional: shuffle the result
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
balanced_df.to_csv("plm.csv" , index=False)
```
